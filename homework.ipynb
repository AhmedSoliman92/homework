```python
class Planets:
    def __init__(self, method, number, orbital_period, mass,distance, year ):
        self.method = method
        self.number = number
        self.orbital_period = orbital_period
        self.mass = mass
        self.distance = distance
        self.year = year 
    def __repr__(self):
        return repr((self.method, self.number, self.orbital_period,self.mass,self.distance, self.year ))
```

```python
import csv

planets_list = []
input_file=csv.DictReader(open("planets.csv"))
for row in input_file:
    planets = Planets(
        method=row['method']
        , number = row['number']
        , orbital_period= row['orbital_period']
        , mass= row['mass']
        , distance = row['distance']
        , year = row ['year']
        )
    planets_list.append(planets)
for i in range(len(planets_list)):
    print(planets_list[i])
```

```python
from datetime import datetime
import numpy as np

for obj in planets_list:
    try:
        int(obj.number)
        if int(obj.number)<=0:
            obj.number= None
    except ValueError:
        obj.number = None
    try:
        if obj.orbital_period is not None:
            obj.orbital_period= float(obj.orbital_period)
            if float(obj.orbital_period)<=0:
                obj.orbital_period= None
    except ValueError:
        obj.orbital_period = None
        
    try:
        if obj.mass is not None:
            obj.mass= float(obj.mass)
            if float(obj.mass)<=0:
                obj.mass= None
    except ValueError:
        obj.mass = None  
    try:
        if obj.distance is not None:
            obj.distance= float(obj.distance)
            if float(obj.distance)<=0:
                obj.distance= None
    except ValueError:
        obj.distance = None
    try:
        int(obj.year)
        if int(obj.year)<=0 or int(obj.year)>= int(datetime.now().strftime("%Y")):
            obj.year= None
    except ValueError:
        obj.year = None

import statistics 
number_list=[]
for i in range(len(planets_list)):
    if planets_list[i].number is None:
        pass
    else:
        number_list.append(int(planets_list[i].number))
median_number = statistics.median(number_list)

for i in range(len(planets_list)):
    if planets_list[i].number is None:
        planets_list[i].number = str(median_number)


orbital_period_list=[]
for i in range(len(planets_list)):
    if planets_list[i].orbital_period is None:
        pass
    else:
        orbital_period_list.append(float(planets_list[i].orbital_period))
median_orbital_period = statistics.median(orbital_period_list)

for i in range(len(planets_list)):
    if planets_list[i].orbital_period is None:
        planets_list[i].orbital_period = str(median_orbital_period)

mass_list=[]
for i in range(len(planets_list)):
    if planets_list[i].mass is None:
        pass
    else:
        mass_list.append(float(planets_list[i].mass))
median_mass = statistics.median(mass_list)

for i in range(len(planets_list)):
    if planets_list[i].mass is None:
        planets_list[i].mass = str(median_mass)


distance_list=[]
for i in range(len(planets_list)):
    if planets_list[i].distance is None:
        pass
    else:
        distance_list.append(float(planets_list[i].distance))
median_distance = statistics.median(distance_list)

for i in range(len(planets_list)):
    if planets_list[i].distance is None:
        planets_list[i].distance = str(median_distance)


year_list=[]
for i in range(len(planets_list)):
    if planets_list[i].year is None:
        pass
    else:
        year_list.append(float(planets_list[i].year))
median_year = statistics.median(year_list)

for i in range(len(planets_list)):
    if planets_list[i].year is None:
        planets_list[i].year = str(median_year)
missing_values = ["n/a", "na", "-", "--", " "]
for obj in planets_list:
    if obj.method in missing_values :
        planets_list.remove(obj)
for i in range(len(planets_list)):
    print(planets_list[i])
```

```python
import pickle

with open('outfile', 'wb') as fp:  #write to file
    pickle.dump(planets_list, fp)

with open ('outfile', 'rb') as fp: #read from file
    itemlist = pickle.load(fp)

print(itemlist)

# Save data into txt file
with open("outfileTXT.csv", "w") as outfile: #write to file
    outfile.write("\n".join(str(item) for item in 
                            planets_list))
    
with open("outfileTXT.csv", "r") as outfile: #read from file
    print(outfile.read())
```
